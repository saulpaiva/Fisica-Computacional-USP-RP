```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='svg'
%matplotlib inline
#site do matplotlib
```

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
s = np.sin(x)
```

```python
plt.plot(x,s,
         label='sin(x)',
         linestyle='-.',
         linewidth=2.5,
         color='red')
plt.xlabel('x')
plt.ylabel('y = sin(x)')
plt.title('Função seno')
plt.legend(loc=3)
plt.xlim(0,10)
plt.grid()
plt.show()
```

```python
x = np.linspace(0,10,50)
```

```python
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
```

```python
plt.plot(x,y,linestyle='',marker='.')
#plt.plot(x,y,linestyle='',marker='o')
plt.errorbar(x,y,yerr=dy,linestyle='',marker='.',ecolor='gray',capsize=5,elinewidth='0.5')
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```

```python
x = np.linspace(-2,2,500)
beta = 40
```

```python
def f(x,beta):
    var = np.exp(-x**2)*np.cos(beta*x)
    return var
```

```python
envoltoria = np.exp(-x**2)
```

```python
plt.plot(x,f(x,beta))
plt.plot(x,envoltoria,color='red')
plt.plot(x,(-1)*envoltoria,color='red')
plt.xlabel('x')
plt.xlabel('f(x)')
plt.xlim(-2,2)
plt.ylim(-1,1)
plt.grid()
plt.show()
```
