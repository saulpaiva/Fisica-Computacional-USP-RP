```python
# Modelo de crescimento logístico
```

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='svg'
%matplotlib inline
```

```python
r = 0.2
k = 50
x = 0.1
dt = 0.01
tempo = np.arange(0,50,dt)
dxdt = []
```

```python
tempo
```

```python
tempo.shape
```

```python
for t in tempo:
    dxdt.append(x)
    x = x + (r*x*(1 - x/k))*dt
```

```python
x
```

```python
dxdt
```

```python
len(dxdt)
```

```python
plt.plot(tempo,dxdt,color='b',lw=1.2)
plt.axhline(k,color='r',ls='--',label='k')
plt.xlim(min(tempo),max(tempo)+dt)
plt.xlabel('Tempo',fontsize=15)
plt.ylabel('x(t)',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(alpha=0.3)
plt.legend(loc=0,fontsize=10)
plt.show()

#plt.savefig('logístico.png',dpi=200)
```

```python
# Lotka-Volterra (presa-predador) em tempo contínuo
```

```python
a = b = 0.7
c = d = 1.0
x = y = 0.1
dt = 0.01
tempo = np.arange(0,50,dt)
dxdt = []
dydt = []
```

```python
for t in tempo:
    dxdt.append(x)
    dydt.append(y)
    x += (a*x - b*x*y)*dt
    y += (d*x*y - c*y)*dt
```

```python
len(dxdt)
```

```python
len(dydt)
```

```python
plt.plot(tempo,dxdt,color='g',ls='-',lw=1.2,label='Presa')
plt.plot(tempo,dydt,color='k',ls='--',lw=1.2,label='Predador')
plt.xlim(min(tempo),max(tempo)+dt)
plt.xlabel('Tempo',fontsize=15)
plt.ylabel('x(t),y(t)',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(alpha=0.3)
plt.legend(loc=0,fontsize=10,facecolor='white')
plt.show()

#plt.savefig('logístico.png',dpi=200)
```

```python
p1 = (0,0)
p2 = (c/d,a/b)
```

```python
plt.plot(dxdt,dydt,color='g',ls='-',lw=1.2)
plt.plot(p1[0],p1[1],ls='',marker='.',color='red')
plt.plot(p2[0],p2[1],ls='',marker='.',color='red')

plt.title('Atrator Ciclo Limite')
plt.ylabel('y(t)',fontsize=15)
plt.xlabel('x(t)',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(alpha=0.3)
#plt.legend(loc=0,fontsize=10,facecolor='white')
plt.show()

#plt.savefig('logístico.png',dpi=200)
```

```python
x0 = np.arange(-2,7,0.2)
y0 = np.arange(-2,7,0.2)
```

```python
X,Y = np.meshgrid(x0,y0)
```

```python
u = (a*X - b*X*Y)*dt
v = (d*X*Y - c*Y)*dt
```
